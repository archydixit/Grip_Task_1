```python
df.groupby(by=["Sub-Category"])["Profit"].sum().reset_index().sort_values(["Profit"]).tail(10).plot(x="Sub-Category",
                                                                                                        y="Profit",
                                                                                                        kind = "bar",
                                                                                                         figsize=(10,5))
plt.show();
```

```python
df.groupby(by=["Category"])["Profit"].sum().reset_index().sort_values(["Profit"]).tail(10).plot(x="Category",
                                                                                                        y="Profit",
                                                                                                        kind = "bar", 
                                                                                                         figsize=(10,5))
plt.show();
```

****Technology is helping earn maximum profit for the company****

```python
df.groupby(by=["State"])["Profit"].sum().reset_index().sort_values(["Profit"]).tail(10).plot(x="State",
                                                                                             y="Profit",
                                                                                             kind = "bar",
                                                                                             figsize=(10,5))
plt.show();
```

**When we compare profit with state, California and New York are giving maximum profit to company**

```python
plt.figure(figsize=(15,5))
sns.barplot(data=df,x='Category',y='Profit',hue='Segment')
plt.show()
```

### ***Now lets Compare Segment with Category and profit, and we can see in above histogram that Technology gives maximum profit under Home Office segment***

```python
plt.figure(figsize=(15,5))
sns.barplot(data=df,x='Region',y='Sales',hue='Category')
plt.show()
```

***If we compare sale in different regions taking into consideration Category, almost all region have more or less similar sales, but in evey region sale in techonological products are maximum. And sale in office supply category is minimum.***

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=df,x='Profit',y='Sales',hue='Region')
plt.show()
```

***East region is in huge loss with very less sales figure and central region are in good profit.***

```python
plt.figure(figsize=(15,5))
sns.barplot(data=df,x='Category',y='Profit',hue='Region')
plt.show()
```

***Profit in Furniture and office supply in the region Centre are very less, with high renge of vaues in both the directions. Where as Profit earned in technology category is highest in east region.***
* Company has to do something about the Furniture category and office supplies in Central region.
* Profit earned in east region under the category Furniture is also has to be worked on.
